   # Summary of Predicting Lung Cancer Survival Time

                                            1) Baseline Hazard

\begin{equation*}
h(t|x) = h_0(t) e^{\beta^t x}
\end{equation*}
         
                                            2) Hazard Ratio

\begin{equation*}
\frac{h(t|x_i)}{h(t|x_j)} = e^{\beta^t x}
\end{equation*}

                                            3) concordance Index
                                            
\begin{equation*}
C_index = \frac{Number of concordant pairs}{Number of admissible pairs}
\end{equation*}

### Understanding definition of Survival function and Hazard function and censorship?
    survival function -> S(t) = P(T < t) 
        where survival function gives us the "probability of survival after time t" and here T is a random lifetime taken from the       population and it should be positive.
    Hazard fucntion :

\begin{equation*}
h(t) = \lim_{\delta t\to 0}\frac{P(t\leq T \leq t | T \le t)}{\delta t}
\end{equation*}

        When deltaT goes to zero implies to the risk of event happening at point of time.
    Censorship :
        Censorship allows us to measure lifetime for the population who haven't experienced the event.
